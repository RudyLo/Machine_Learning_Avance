# Part 1: Manipulating Data with Pandas library

```python
import pandas as pd

#Import des données avec selection des colonnes nécessaires
film_df = pd.read_csv("movies_metadata.csv", usecols= ["title","release_date","budget","revenue","runtime","genres"] )
film_df.head()
```

```python
film_df.dtypes
```

```python
float_budget = float('budget')
```

Il n'est pas possible de convertir un string en float

```python
def to_float(number):
    for i in range (len(number)):
        if type(number) == 'str':
            number[i] == 'NA'
        else :
            number[i]=float(number[i])
            
    afficher=to_float(number)
    
    print(afficher)
```

```python
def float_test(column):
    for i in range (len(column)) :
        
        column[i] = pd.to_numeric(column, downcast='signed')
        
    afficher=float_test(column)
    
    print(afficher)
```

```python
#Conversion des la colonne budget en float
film_df['budget'] = pd.to_numeric(film_df['budget'], errors='coerce')
```

```python
#Création de la colonne year dans le dataframe
film_df['year'] = pd.to_datetime(film_df['release_date'], errors='coerce')

#On extrait year dans le datetime
film_df['year'] = pd.DatetimeIndex(film_df['year']).year



print(film_df.dtypes)

#Affichage du Dataframe
film_df
```

```python
#Affiche du film le plus vieux
film_df[ film_df['year'] == film_df['year'].min() ]
```

Le plus vieux film est le Passage de Venus réalisé en 1874

```python
#Affichage du revenu max
film_df[ film_df['revenue'] == film_df['revenue'].max() ]
```

Le film le plus rentable est Avatar 

```python
#On créer un dataframe avec un revenu supérieur à 1 milliard
df_mask=film_df['revenue']>=1000000000
filtered_df = film_df[df_mask]
filtered_df
```

```python
#On créer un dataframe avec un revenu supérieur à 1 milliard et un budget inférieur à 150 millions
df_mask=filtered_df['budget']<=150000000
filtered_df2 = filtered_df[df_mask]
filtered_df2
```
